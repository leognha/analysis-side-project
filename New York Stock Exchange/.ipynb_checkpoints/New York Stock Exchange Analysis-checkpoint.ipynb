```python
nRowsRead = 1000 # 先讀1000行起來看
df1 = pd.read_csv('data/fundamentals.csv', delimiter=',', nrows = nRowsRead)
df1.dataframeName = 'fundamentals.csv'
nRow, nCol = df1.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
df1.head(5) # 列頭五比來看
```

```python
# 分佈圖（直方圖/條形圖）的列數據
def plotPerColumnDistribution(df, nGraphShown, nGraphPerRow):
    nunique = df.nunique() # 計算每列的唯一值數量
    df = df[[col for col in df if nunique[col] > 1 and nunique[col] < 50]] # 篩選出唯一值數量在1和50之間的列
    nRow, nCol = df.shape
    columnNames = list(df)
    # 計算每行的圖表數量
    nGraphRow = int((nCol + nGraphPerRow - 1) / nGraphPerRow)
    # 設置圖表大小
    plt.figure(num = None, figsize = (6 * nGraphPerRow, 8 * nGraphRow), dpi = 80, facecolor = 'w', edgecolor = 'k')
    # 繪製圖表
    for i in range(min(nCol, nGraphShown)):
        print(nCol, nGraphShown)
        plt.subplot(nGraphRow, nGraphPerRow, i + 1)
        columnDf = df.iloc[:, i]
        # 如果列數據不是數字，則繪製條形圖
        if (not np.issubdtype(type(columnDf.iloc[0]), np.number)):
            valueCounts = columnDf.value_counts()
            valueCounts.plot.bar()
        else:
            columnDf.hist()
        plt.ylabel('counts')
        plt.xticks(rotation = 90)
        plt.title(f'{columnNames[i]} (column {i})')
    plt.tight_layout(pad = 1.0, w_pad = 1.0, h_pad = 1.0)
    plt.show()
```

```python
plotPerColumnDistribution(df1, 10, 5)
```

```python
# Correlation matrix
def plotCorrelationMatrix(df, graphWidth):
    filename = df.dataframeName
    df = df.dropna('columns') # 將所有包含NaN的列刪除
    df = df[[col for col in df if df[col].nunique() > 1]] # 刪除所有唯一值的列
    # 如果df的列數小於2，則不顯示相關性圖
    if df.shape[1] < 2:
        print(f'沒有顯示相關性圖：非NaN或常數列的數量（{df.shape[1]}）小於2')
        return
    # 繪製相關性圖
    corr = df.corr()
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k')
    corrMat = plt.matshow(corr, fignum = 1)
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.gca().xaxis.tick_bottom()
    plt.colorbar(corrMat)
    plt.title(f'Correlation Matrix - {filename}', fontsize=15)
    plt.show()
```

```python
# Correlation matrix:
plotCorrelationMatrix(df1, 19) # 看特徵相關性
```

```python
# 散佈 & 密度圖
def plotScatterMatrix(df, plotSize, textSize):
    df = df.select_dtypes(include =[np.number]) # 只選擇數值變數
    # 刪除缺失值所在的行列
    df = df.dropna('columns')
    df = df[[col for col in df if df[col].nunique() > 1]] # 刪除只有一個值的變數
    columnNames = list(df) 
    if len(columnNames) > 10: # 限制最多 10 個變數
        columnNames = columnNames[:10]
    df = df[columnNames]
    # 繪製散點圖
    ax = pd.plotting.scatter_matrix(df, alpha=0.75, figsize=[plotSize, plotSize], diagonal='kde')
    corrs = df.corr().values
    # 在圖上標註相關係數 (np.triu_indices_from 獲取上三角矩陣的索引)
    for i, j in zip(*plt.np.triu_indices_from(ax, k = 1)):
        ax[i, j].annotate('Corr. coef = %.3f' % corrs[i, j], (0.8, 0.2), xycoords='axes fraction', ha='center', va='center', size=textSize)
    plt.suptitle('Scatter and Density Plot')
    plt.show()
```

```python
# 散佈 & 密度圖
# 1.顯示變量間的關係：散佈圖可以用來觀察兩個連續變量之間的關係。例如，可以幫助識別變量之間是否存在線性或非線性的關係。
# 2.檢測異常值：散佈圖可以很容易地顯示出數據中的異常值，這些點通常會遠離其他數據點。
# 3.趨勢識別：通過散佈圖，可以初步識別出數據中的趨勢，例如正相關、負相關或無關。
plotScatterMatrix(df1, 20, 10)
```

### prices.csv

```python
nRowsRead = 1000 # specify 'None' if want to read whole file
# prices.csv may have more rows in reality, but we are only loading/previewing the first 1000 rows
df2 = pd.read_csv('data/prices.csv', delimiter=',', nrows = nRowsRead)
df2.dataframeName = 'prices.csv'
nRow, nCol = df2.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
df2.head(5)
```

```python
# Distribution graphs (histogram/bar graph) of sampled columns:
plotPerColumnDistribution(df2, 10, 5)
```

```python
# Correlation matrix:
plotCorrelationMatrix(df2, 8)
```

```python
# 換用 seaborn 來標註相關係數
corr = df2.corr()
# 繪製相關矩陣
plt.figure(figsize=(10, 8))
# annot=True：在每個單元格中標註數值。
ax = sns.heatmap(corr, annot=True, fmt=".2f", cmap='viridis', vmin=0, vmax=1)

# 設置標題
plt.title('Correlation Matrix for prices.csv')
plt.show()
```
